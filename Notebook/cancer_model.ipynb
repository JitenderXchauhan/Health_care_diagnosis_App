```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("lung_cancer.csv")
```

```python
data.head()
```

```python
data['LUNG_CANCER'].value_counts()
```

```python
data.groupby('LUNG_CANCER').mean()
```

```python
X = data.drop(columns = 'LUNG_CANCER', axis=1)
Y = data['LUNG_CANCER']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

Accuracy Score

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

Function to Evaluate the Disease Level

```python
def classify_disease_level(probability):
    if probability < 0.4:
        return "Low Risk"
    elif 0.4 <= probability < 0.7:
        return "Moderate Risk"
    else:
        return "High Risk"
```

```python
test_probabilities = classifier.predict_proba(X_test)[:, 1]  # Probabilities for class '1'
test_disease_levels = [classify_disease_level(prob) for prob in test_probabilities]
```

```python
results_df = pd.DataFrame({

    "Predicted Probability": test_probabilities,
    "Disease Level": test_disease_levels
})
```

```python
print("\nDisease Levels for Test Set:")
print(results_df)
```
